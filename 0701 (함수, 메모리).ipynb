```python
# list comprehension

myList = [1, 2, 3, 4, 5, 6, 7, 8, 9]

# (문제) 위 리스트에서 각각 요소에 *2를 해서 새로운 list를 만들자!

# 방법 1
result = list()
for a in myList:
    result.append(a*2)
print(result)

# 방법 2 - List Comprehension
result = [a*2 for a in myList]
print(result)


# (문제) 위 리스트에서 짝수 요소에 *2를 해서 새로운 list를 만들자!
result = [a*2 for a in myList if a % 2 == 0]
print(result)
```

### 메모리

파이썬은 일급함수(1st classes function)를 지원한다

```python
a = 100
print(a)   # 100이라고 저장되어있는 것을 a가 가르켜서, 이를 프린트한 것

# a라는 변수가 가지고 있는 주소값을 찍고 싶다 
# ---> 내장함수인 id를 이용!
print(id(a))   # 94769906861664 ---> 100이라는 정수값이 저장된 메모리 주소값


a = [1, 2, 3]
print(id(a))   # 140692373856080
b = [1, 2, 3]
print(id(b))   # 140692503316992

print(a == b)   # True ---> 내용이 같은지
print(id(a) == id(b))   # False ---> 주소값이 같은지


a = 50
b = 50
print(a is b)   # True ---> 50이 하나밖에 없고 a와 b가 동시에 바라보고 있다

# 왜??? 
# 숫자는 워낙 많이 사용되어서, 256까지는 인스턴스(객체)를 한개만 만들어 공유한다
# 메모리 효율성!

a = 256
b = 256
print(a is b)   # True

a = 257
b = 257
print(a is b)   # False
```

Python의 함수는 default parameter를 사용할 수 있다
- 맨 끝에만 설정할 수 있음

```python
def my_sum(a, b, c=True):
    if c:
        print(a + b)
    else:
        print(a * b)

my_sum(10, 20, True)
my_sum(10, 20)   # default parameter는 입력인자를 생략할 수 있다!
```

python 함수는 입력인자에 대해 mutable, immutable 특성이 있다.
- mutable (가변성의) : 
    - call-by-value
    - 리스트는 함수의 인자로 넘어갔을 때 변함
- immutable (불변성의) : 
    - call-by-reference
    - 숫자, 문자열, 튜플은 함수의 인자로 넘어갔을 때 불변함

```python
def my_func(tmp_number, tmp_list):
    tmp_number += 10
    tmp_list.append(100)

data_x = 10
data_list = [1, 2, 3]

print('현재 data_x : {}, data_list : {}'.format(data_x, data_list))

my_func(data_x, data_list)
print('호출 후 data_x : {}, data_list : {}'.format(data_x, data_list))

# 결과: 
# data_x는 10으로 변하지 않았고 ---> immutable
# data_list는 [1, 2, 3, 100]으로 변했다 ---> mutable
```

> ### Local variable Vs. Global variable
- global 변수 별로 안씀 $\rightarrow$ 유지보수 힘듦

```python
tmp = 100

def my_func(x):
    tmp += 10
    return tmp

print(my_func(100)) 

# ---> 오류남. 왜??? scope에 문제가 있음
# 함수 안의 tmp는 지역변수로, 밖에서 선언된 글로벌변수 tmp를 못찾음! 
```

```python
tmp = 100

def my_func(x):
    global tmp   # 밖에서 찾아라~
    tmp += 10
    return tmp

print(my_func(100))
```

```python
# 하지만 global var를 잘 쓰지는 않는다.
# 글로벌은 필요한 경우에만 쓰자!

def my_func(x):
    tmp = 100
    tmp += 10
    return tmp

print(my_func(100))
```

> ### sort vs. sorted

- sort는 method로 주체에 적용되어 원본이 변함. 리턴이 없음!
- sorted는 method가 아닌 함수(안에 들어오는 인자에 대해 특정 작업을 하고 리턴을 줄수도, 리턴을 안줄수도)로 리턴을 주는 형태의 함수다. 

```python
### sort

my_list = [3, 6, 8, 1, 2, 7]
my_list.sort()   # return이 없고, 원본이 변함
print(my_list)


### sorted

my_list = [3, 6, 8, 1, 2, 7]
sorted(my_list)
print(my_list)   # 원본이 변하지 않음!
```
