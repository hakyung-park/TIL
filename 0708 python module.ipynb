# Module
- module이란? 변수, function(함수), class를 저장하고 있는 파일
- 하나의 python 파일 : module1.py
- 현재 우리가 사용하는 파일 : .ipynb ---> jupyter notebook파일로, cell로 구성
- module을 사용하는 이유 : 코드의 재 사용성, 관리
- import를 이용하여 module을 불러오기
- python의 모든 것은 객체. 따라서 module도 객체로 인식
- import할 때 as 로 별명을 만들어서 사용할 수 있다.
- python의 module은 크게 2 종류가 있다.
    1. C 언어로 구현된 module
    2. Python 언어로 구현된 module

        - 윈도우 시스템의 환경변수를 하나 만들자 ---> PYTHONPATH (module을 모아놓을 폴더) ---> DB.md 파일 참고

    - 동적으로 추가한 환경 변수를 알지 못함 ---> 사용하는 프로그램을 재시작해야 함

- module은 외장 module, 내장 module로 구분된다.
    - 내장 module : python과 함께 설치되는 module. (따로 다운로드 할 필요 없음)
    - 외장 module : repository 안에 모듈이 들어가 있고, 그 것을 다운로드해서 사용해야 하는 module

```python
### (시스템적으로 설정 후) 모듈 불러들여서 변수, 함수, class를 이용하자

import module1 as m1   # ERROR! ---> jupyter notebook으로는 됐을 것! 로컬 설정은 해놓았기 때문에

print(m1.my_PI)
print(m1.my_func(100,200))
car1 = m1.MyCar('현대', 3000)
```

```python
### 선택적으로 모듈을 불러들이기

# from 모듈이름/패키지이름 import 함수/변수/class 

from module1 import my_PI   # module1에서 my_PI 변수만 가져오자
from module1 import *       # module1에 있는 모든 것을 다 가져오자 
```

```python
### Q : import module1 과 from module1 import *  차이점

# import 뒤에 나오는 것들은 무조건 module1.~ 항상 써야함

# 따라서, 
import module1           # 는 module1.my_PI 
from module1 import *    # 는 my_PI 
```

# Package 
- module은 계층적으로 관리하는 게 좋다.
- 폴더를 계층화시켜 특정 module을 역할에 맞는 폴더에 넣어서 관리
- 이를 Package라고 부른다. 즉, module을 담고 있는 바구니!
- python에서 . 을 이용한다. 이 때 패키지의 계층구조를 나타내는 의미다. python module을 관리
    - import A.B
    - A는 폴더명 (패키지명)
    - B는 폴더 A 안에 들어있는 모듈

```python
# 설정으로 C:\python_lib 으로 잡혀있는 상태
# 그 안에 myFolder라는 폴더 생성, 그 안에 myModule 폴더 생성, 그 안에 module1.py 생성

import myFolder.myModule.module1 as m1

print(m1.my_PI)
```

```python
# 거의 모든 모듈은 package화 되어있기 때문에 명시해야 하지만 from절을 이용해서 다음과 같이 쓰는게 보편적인 사용 방법이다. 
from myFolder.myModule import module1

print(module1.my_PI)
```
